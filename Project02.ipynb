### Với mỗi cột có kiểu dữ liệu dạng số (numerical), các giá trị được phân bố như thế nào

```python
tsunami_df.describe().T
```

Có vẻ như các dữ liệu dạng số ở phần nửa trên tương đối đầy đủ so với nửa dưới.

#### Số-lượng/tỉ-lệ các giá trị thiếu?

```python
nume_col_info_df = pd.DataFrame()
for col in tsunami_df.keys():
    if tsunami_df[col].dtype in [np.float32, np.float64]:
        missing_percentage = round(tsunami_df[col].isnull().sum() * 100 / len(tsunami_df[col]), 1)
        min_value = tsunami_df[col].min()
        lower_quartile = tsunami_df[col].quantile(0.25)
        median = tsunami_df[col].median()
        upper = tsunami_df[col].quantile(0.75)
        max_value = tsunami_df[col].max()
        row_line = pd.Series([missing_percentage, min_value, lower_quartile, median, upper, max_value], 
                             index = ['missing_percentage', 'min', 'lower_quartile', 'median', 'upper_quartile', 'max'])
        nume_col_info_df[col] = row_line
nume_col_info_df.T
```

#### Min? max? → Có gì bất thường không?

Dựa vào dataframe ở trên thì có vẻ như min max không có gì bất thường ở dữ liệu này.

### Với mỗi cột có kiểu dữ liệu dạng phân loại (categorical), các giá trị được phân bố như thế nào?

#### Số-lượng/tỉ-lệ các giá trị thiếu?

```python
cate_col_info_df = pd.DataFrame()
for col in tsunami_df.keys():
    if tsunami_df[col].dtype not in [np.float32, np.float64]:
        missing_percentage = round(tsunami_df[col].isnull().sum() * 100 / len(tsunami_df[col]), 1)
#         if col in multichoice_cols:
#             x = tsunami_df[col].str.split(';')
#             num_values = len(x.explode().value_counts())
#             value_percentages = round(x.explode().value_counts() * 100/x.explode().value_counts().sum(), 1)
#         else:
        num_values = len(tsunami_df[col].value_counts())
        value_percentages = round(tsunami_df[col].value_counts() * 100/tsunami_df[col].notnull().sum(), 1)
        cate_col_info_df[col] = pd.Series([missing_percentage, num_values, value_percentages], 
                                          index=['missing_percentage', 'num_values', 'value_percentages'])
cate_col_info_df
```

#### Số lượng các giá trị khác nhau? Show một vài giá trị → Có gì bất thường không?

```python
country_df = pd.DataFrame(cate_col_info_df['Country']['value_percentages'])
country_df.head(5)
```

```python
location_df = pd.DataFrame(cate_col_info_df['Location Name']['value_percentages'])
location_df.head(5)
```
